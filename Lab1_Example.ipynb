```python
#Using matplotlib in python compare the Coffee product, Salse and Profit

#1 line graph with proper legends (Product vs Salse,Product vs Profit)
#1 barplot showing Product vs Salse vs Profit
#1 boxplot between Type and Profit
#1 pie chart to find which type is more profitable

# What conclusion can you make from these plots?
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Coffee= pd.read_csv('/content/drive/MyDrive/IMSE_441/Coffee.csv')
Coffee
```

```python
plt.figure(figsize=(15,10))
plt.plot(Coffee['Product'],Coffee['Sales'], label='Sales')
plt.plot(Coffee['Product'],Coffee['Profit'], color='red', label= 'Profit')
plt.title('Coffee sales vs profit' )
plt.xlabel('Product')
plt.ylabel('Money')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.bar(Coffee['Product'],Coffee['Sales'], label='Sales')
plt.bar(Coffee['Product'],Coffee['Profit'], color='red',bottom=Coffee['Sales'], label= 'Profit')
plt.title('Coffee sales vs profit' )
plt.xlabel('Product')
plt.ylabel('Money')
plt.legend()
plt.show()
```

```python
Regular=(Coffee[Coffee['Type']=='Regular']['Profit'])
Regular 
```

```python
Decaf=(Coffee[Coffee['Type']=='Decaf']['Profit'])
Decaf
```

```python
plt.figure(figsize=(15,10))
plt.boxplot([Regular,Decaf])
plt.title('Coffee Regual vs Decaf' )
plt.xlabel('Type')
plt.ylabel('Profit')
plt.show()
```

```python
Type=['Regular','Decaf']
Profit=[sum(Regular),sum(Decaf)]
plt.figure(figsize=(8,8))
plt.pie(Profit,labels=Type,autopct='%0.2f%%', explode=[0,0.1])
plt.title('Coffee Regual vs Decaf' )
plt.show()
```

# **Conclusions**

On average Decaf coffee are expensive than Regular coffee but the boxplot is saying Decaf has higher profit margin than Regular. From Piechart We are seeing Both type has equal profit share overall.
