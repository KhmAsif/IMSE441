```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

**Ploting Normal distribution**

**Finding the probability**

```python
#Finding the normal PDF
#norm.pdf(X,mean,std)
PDF_X=norm.pdf(13,10,2)
PDF_X
```

```python
#Finding the normal CDF
#norm.cdf(X, mean , sd).cdf(X)
CDF_X=1- norm.cdf(13,10,2)
CDF_X
```

```python
#Finding the normal CDF for range
CDF_xu= norm.cdf(-1.25,0,1)
CDF_xl= norm.cdf(0.37,0,1)
CDF_xr=CDF_xu-CDF_xl
CDF_xr
```

**Finding the z value**

```python
#Finding the z value for Standard normal random variable
z_1= norm.ppf(0.99,0,1)
z_1
```

```python
#Finding the z value for normal random variable
z_1= norm.ppf(0.85,150,3.8)
z_1
```

```python
#Finding mean and variance of normal distribution
x = np.random.randn(20)
mu, sigma  = np.mean(x), np.std(x)
z_2= norm.ppf(0.85,mu,sigma)
z_2
```

**Ploting of Normal Distribution**

```python
#PDF
x=np.arange(-1,5,0.1)
plt.plot(x,norm.pdf(x,2,1))
plt.show()
```

```python
#Standard Normal PDF
x=np.arange(-3,3,0.01)
plt.plot(x,norm.pdf(x,0,1))
plt.show()
```

```python
#CDF
plt.plot(x,norm.cdf(x,0,1))
plt.show()
```

**Checking for normality**

```python
from seaborn.regression import statsmodels
import seaborn as sns
import scipy.stats as stats
```

```python
Normal= np.random.randn(1000)
Skewed= stats.skewnorm.rvs(a=5,size=1000)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 7))
sns.histplot(Normal,kde=True, color ='blue',ax=ax[0])
stats.probplot(Normal, plot=plt)
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 7))
sns.histplot(Skewed,kde=True, color ='g',ax=ax[0])
stats.probplot(Skewed, plot=plt)
plt.show()
```
